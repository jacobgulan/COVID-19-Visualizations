```python
# importing packages
import pandas as pd
import numpy as np
```

```python
# importing data
path = "/content/drive/MyDrive/4122 Visual Analytics/"
task = pd.read_csv(path + "task3.csv")
```

```python
# creating population column
task['population'] = ''

for index, row in task.iterrows():
  fv = task['fully_vaxed'][index]
  pfv = task['%_fully_vaxed'][index]
  pfv = pfv/100
  task['population'][index] = fv/pfv

task
```

```python
# creating covid% column
task['covid%'] = ''

for index, row in task.iterrows():
  c = task['covid_cases'][index]
  p = task['population'][index]
  task['covid%'][index] = (c/p)*100

task
```

```python
# removing extra clumns
task.drop(columns=task.columns[0], axis=1, inplace=True)

# removing nulls
task.dropna(inplace=True)

task
```
